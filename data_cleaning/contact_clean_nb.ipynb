```python
import pandas as pd
```

```python
contact_path = "../data/Contact_details.csv"
contact_df = pd.read_csv(contact_path)
contact_df.head()
```

```python
country_map = {
    "United States": "US",
    "America": "US",
    "United Kingdom": "UK",
    "Uk": "UK"
}

contact_df["Country"] = contact_df["Country"] \
    .map(lambda x: x if x not in country_map.keys() else country_map[x])
```

```python
contact_df.Country.value_counts()
```

```python
contact_df.drop(["Contact Name", "Local Salutation"],
                axis=1,
                inplace=True
               )
```

```python
contact_df
```

```python
clean_contact_path = "../cleaned_data/contact_details_cleaned.csv"
contact_df.to_csv(clean_contact_path, index=False)
```
